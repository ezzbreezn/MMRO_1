В соревновании в качестве метрики качества использовалось RMSLE:
$$\text{RMSLE}(X, y, a) = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2}$$

**Вопрос (0.1 балл)**: **(?)**Как вы думаете, почему авторы соревнования выбрали именно RMSLE, а не RMSE?

Можно сказать, что по сути RMSLE - это RMSE, посчитанная в логарифмическом масштабе. При использовании RMSE при наличии выбросов, то есть объектов, у которых значения некоторых признаков сильно отличаются от всех других объектов, значение ошибки будет достаточно велико. При использовании RMSLE значения признаков будут отмасштабированы по логарифмической шкале, в случае присутствия выбросов ошибка будет меньше, чем у RMSE, в силу свойств логарифма (функция выпукла, рост функции замедляется с ростом аргумента), влияние ошибки на выбросе будет меньше. Таким образом, RMSLE более устойчива к выбросам, чем RMSE, вклад ошибок на выбросах меньше. В решаемой задаче как раз присутствуют выбросы. Также можно заметить, что если прогноз меньше истинного значения или больше на одну и ту же величину, то в случае недооценки штраф будет больше, чем в случае переоценки (так же в силу свойств логарифма), т. е. RMSLE сильнее штрафует недопрогноз. Это может согласовываться с решаемой задачей, так как целевой признак - время поездки, и чем оно дольше, тем больше стоимость поездки, и в таком случае выгоднее будет переоценить.

На семинаре мы рассматривали несколько моделей линейной регрессии в `sklearn`, но каждая из них оптимизировала среднеквадратичную ошибку (MSE), а не RMSLE. Давайте проделаем следующий трюк: будем предсказывать не целевую переменную, а ее *логарифм*. Обозначим $\hat{y}_i = \log{(y_i + 1)}$ — модифицированный таргет, а $\hat{a}(x_i)$ — предсказание модели, которая обучалась на $\hat{y}_i$, то есть логарифм таргета. Чтобы предсказать исходное значение, мы можем просто взять экспоненту от нашего предсказания: $a(x_i) = \exp(\hat{a}(x_i)) - 1$.

**Вопрос (0.3 балла)**: **(?)**Покажите, что оптимизация RMSLE для модели $a$ эквивалентна оптимизации MSE для модели $\hat{a}$.

**Доказательство**: Напишие прямо тут в ячейке, при необходимости использовав синтаксис LaTeX в ячейке

╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ

Чтобы показать эквивалентность оптимизаций, покажем, что соответствующие метрики имеют одинаковые точки минимума.

$$\text{RMSLE}(X, y, a) = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2}$$

$$\text{MSE}(X, \hat{y}, \hat{a}) = \frac{1}{\ell}\sum_{i=1}^{\ell}(\hat{y_{i}} - \hat{a}(x_{i}))^2  = \frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2$$

Очевидно, что минимальное указанное значение обеих метрик равно 0, достигается оно в том случае, когда каждое из слагаемых под знаком суммы равно 0. В таком случае $ \log{(y_i+1)} = \log{(a({x_i})+1)}$  $ \forall i=\overline{1,\ell} $. Значения $y_i$ известны и определены однозначно, логарифм является строго монотонной функцией, следовательно, значения $a(x_i)$, обеспечивающие минимум, также определяются однозначно: $a(x_i)=y_i$ $\forall i=\overline{1,\ell}$. Таким образом, точки минимума метрик совпадают, следовательно, их оптимизации эквивалентны.

Мы свели задачу оптимизации RMSLE к задаче оптимизации MSE - квадратичной функции потерь. Для нее в данном случае лучшим константным предсказанием будет выборочное среднее $\frac{1}{\ell}\sum_{i=1}^{\ell}\log(y_i+1)$, то есть среднее значение признака log_trip_duration

Обратите внимание, что в эту функцию мы передаем лишь небольшой кусочек данных, посколько иначе функция будет работать очень долго

```python
def show_circles_on_map(data, latitude_column, longitude_column, color):
    """
    The function draws map with circles on it.
    The center of the map is the mean of coordinates passed in data.
    
    data: DataFrame that contains columns latitude_column and longitude_column
    latitude_column: string, the name of column for latitude coordinates
    longitude_column: string, the name of column for longitude coordinates
    color: string, the color of circles to be drawn
    """

    location = (data[latitude_column].mean(), data[longitude_column].mean())
    m = folium.Map(location=location)

    for _, row in data.iterrows():
        folium.Circle(
            radius=100,
            location=(row[latitude_column], row[longitude_column]),
            color=color,
            fill_color=color,
            fill=True
        ).add_to(m)

    return m
```

Мы не можем посчитать точное расстояние, которое необходимо преодолеть такси, но мы можем его оценить, посчитав кратчайшее расстояние между точками начала и конца поездки. Чтобы корректно посчитать расстояние между двумя точками на Земле, можно использовать функцию `haversine`. Посчитайте кратчайшее расстояние для объектов и запишите его в колонку `haversine`:

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
from sklearn.metrics.pairwise import haversine_distances
from math import radians
X_train['haversine'] = X_train.apply(lambda x: (haversine_distances([[radians(x['pickup_latitude']), radians(x['pickup_longitude'])],
                                                                    [radians(x['dropoff_latitude']), radians(x['dropoff_longitude'])]]) * 6371)[0, 1], axis=1)
```

```python
X_test['haversine'] = X_test.apply(lambda x: (haversine_distances([[radians(x['pickup_latitude']), radians(x['pickup_longitude'])],
                                                                    [radians(x['dropoff_latitude']), radians(x['dropoff_longitude'])]]) * 6371)[0, 1], axis=1)
```

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
X_train['log_haversine'] = np.log1p(X_train['haversine'])
X_test['log_haversine'] = np.log1p(X_test['haversine'])
```

**Задание 9 (1 балл)**. Сейчас мы почти что не используем сами значения координат. На это есть несколько причин: по отдельности рассматривать широту и долготу не имеет особого смысла, стоит рассматривать их вместе. Во-вторых, понятно, что зависимость между нашим таргетом и координатами не линейная. Чтобы как-то использовать координаты, можно прибегнуть к следующему трюку: обрамим область с наибольшим количеством поездок прямоугольником (как на рисунке). Разобьем этот прямоугольник на ячейки. Каждой точке сопоставим номер ее ячейки, а тем точкам, что не попали ни в одну из ячеек, сопоставим значение -1.

Напишите трансформер, который сначала разбивает показанную на рисунке область на ячейки, а затем создает два признака: номер ячейки, в которой началась поездка, и номер ячейки, в которой закончилась поездка. Количество строк и столбцов выберите самостоятельно.

Обратите внимание, что все вычисления должны быть векторизованными, трансформер не должен модифицировать передаваемую ему выборку inplace, а все необходимые статистики (если они вдруг нужны) нужно считать только по обучающей выборке в методе `fit`:

```python
from sklearn.base import BaseEstimator, TransformerMixin

class MapGridTransformer(BaseEstimator, TransformerMixin):
    #╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
    def __init__(self, rows=5, columns=5):
        self.rows = rows
        self.columns = columns

    def fit(self, X):
        p_lat_mean = X['pickup_latitude'].mean()
        p_lat_var = X['pickup_latitude'].var()
        p_long_mean = X['pickup_longitude'].mean()
        p_long_var = X['pickup_longitude'].var()
        d_lat_mean = X['dropoff_latitude'].mean()
        d_lat_var = X['dropoff_latitude'].var()
        d_long_mean = X['dropoff_longitude'].mean()
        d_long_var = X['dropoff_longitude'].var()
        self.lat_mean = (p_lat_mean + d_lat_mean) / 2
        self.long_mean = (p_long_mean + d_long_mean) / 2
        self.lat_std = np.sqrt(p_lat_var + d_lat_var) / 2
        self.long_std = np.sqrt(p_long_var + d_long_var) / 2
        self.lat_step = 2 * self.lat_std / self.rows
        self.long_step = 2 * self.long_std / self.columns
        return self

    def transform(self, X):
        cond1 = (X['pickup_latitude'] >= self.lat_mean - self.lat_std) & (X['pickup_latitude'] <= self.lat_mean + self.lat_std) & (X['pickup_longitude'] >= self.long_mean - self.long_std) & (X['pickup_longitude'] <= self.long_mean + self.long_std)
        cond2 = (X['dropoff_latitude'] >= self.lat_mean - self.lat_std) & (X['dropoff_latitude'] <= self.lat_mean + self.lat_std) & (X['dropoff_longitude'] >= self.long_mean - self.long_std) & (X['dropoff_longitude'] <= self.long_mean + self.long_std)
        X_trans = X.copy(deep=True)
        X_trans['start_cell'] = np.where(cond1, (self.lat_mean + self.lat_std - X['pickup_latitude']) // self.lat_step * self.columns + (X['pickup_longitude'] - self.long_mean + self.long_std ) // self.long_step + 1, -1)
        X_trans['end_cell'] = np.where(cond2, (self.lat_mean + self.lat_std - X['dropoff_latitude']) // self.lat_step * self.columns + (X['dropoff_longitude'] - self.long_mean + self.long_std ) // self.long_step + 1, -1)
        return X_trans
```

**(0.5 балла)** Разбейте обучающую выборку на обучающую и валидационную в отношении 8:2. По валидационной выборке подберите оптимальные значения параметра регуляризации (по логарифмической сетке) отдельно для `Ridge` и отдельно для `Lasso`, на тестовой выборке измерьте качество лучшей полученной модели.

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
X_t, X_v, y_t, y_v = train_test_split(X_ohe, y_train_filtered, test_size=0.2, random_state=0)
alphas = np.logspace(-5, 0, 100)

fst = False
best_lasso_mse = 0
best_ridge_mse = 0
best_lasso_alpha = 0
best_ridge_alpha = 0

for alpha in alphas:
    print(f'alpha: {alpha}')
    clf1 = Ridge(alpha)
    clf2 = Lasso(alpha)
    clf1.fit(X_t.values, y_t.values)
    clf2.fit(X_t.values, y_t.values)
    res1 = clf1.predict(X_v.values)
    res2 = clf2.predict(X_v.values)

    mse_ridge = np.sum((y_v.values - res1) ** 2) / y_v.shape[0]
    mse_lasso = np.sum((y_v.values - res2) ** 2) / y_v.shape[0]
    print(f'Ridge MSE: {mse_ridge}')
    print(f'Lasso MSE: {mse_lasso}')
    if not fst:
        best_lasso_mse = mse_lasso
        best_ridge_mse = mse_ridge
        best_lasso_alpha = alpha
        best_ridge_alpha = alpha
        fst = True
    else:
        if mse_lasso < best_lasso_mse:
            best_lasso_mse = mse_lasso
            best_lasso_alpha = alpha
        if mse_ridge < best_ridge_mse:
            best_ridge_mse = mse_ridge
            best_ridge_alpha = alpha

print(f'Best Ridge alpha: {best_ridge_alpha}, MSE: {best_ridge_mse}')
print(f'Best Lasso alpha: {best_lasso_alpha}, MSE: {best_lasso_mse}')
```
